```python
%load_ext autoreload
%autoreload 2

import ipywidgets as _ipyw
import time

from app import initialize_ui
from gui_components import display_guide, get_file_input_css_style
from IPython.display import HTML, clear_output as _clear_cell, display

from hysprint_utils.access_token import get_token, log_notebook_usage
from hysprint_utils.api_calls import (
    get_batch_ids,
    get_ids_in_batch,
    get_nomad_ids_of_entry,
    get_sample_description,
)
from hysprint_utils.config import API_ENDPOINT, URL_BASE

URL_API = URL_BASE + API_ENDPOINT

# Clear any saved or previously accumulated cell outputs.
# VS Code Jupyter does not clear cell outputs on re-run; without this, each run
# (and each notebook open) adds a new copy of all widgets below the old ones.
# wait=True means the clear fires on the first new output below, avoiding a
# blank-screen flash.
_clear_cell(wait=True)
display(HTML("<script>document.title='NOMAD File Uploader'</script>"))
log_notebook_usage()
display(HTML(get_file_input_css_style()))

TOKEN = get_token(URL_API)

print("Getting batch IDs...")
try:
    batch_ids_list_tmp = list(get_batch_ids(URL_API, TOKEN))
    time.sleep(0.2)
    batch_ids_list = []
    for b in batch_ids_list_tmp:
        if "_".join(b.split("_")[:-1]) in batch_ids_list_tmp:
            continue
        batch_ids_list.append(b)
except Exception as _e:
    print(f"Could not fetch batch IDs ({_e.__class__.__name__}: {_e}). Starting with empty list.")
    batch_ids_list = []

print("Initialising UI...")
try:
    initialize_ui(
        batch_ids_list,
        get_batch_ids,
        get_ids_in_batch,
        get_sample_description,
        get_nomad_ids_of_entry,
    )
except Exception as _init_err:
    import traceback as _tb
    print(f"UI initialisation failed: {_init_err}")
    _tb.print_exc()
```
